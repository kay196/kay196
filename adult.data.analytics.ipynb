```python
acb=df['native-country'].value_counts()
country_list=[]
for i in acb.index:
   country_list.append(i)
```

```python
highest_per=0
for country in country_list:
    a= df[(df['native-country']==country)& (df['salary']=='>50K')].shape[0]/df[df['native-country']==country].shape[0]*100
    if a>highest_per:
        highest_per=a
        country_highest_per=country
print('highest percentage is', highest_per ,'and the country with highest percentage is' , country_highest_per)
```
