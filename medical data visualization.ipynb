```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pylab import savefig
```

```python
# Import data
df = pd.read_csv('medical_examination.csv')
```

```python
# Add 'overweight' column
df['BMI'] = df['weight']/((df['height']/100)**2)
df['overweight']=[1 if x>25 else 0 for x in df['BMI']] #overweight= 1, else =0
```

```python
# Normalize data by making 0 always good and 1 always bad. If the value of 'cholesterol' or 'gluc' is 1, make the value 0. If the value is more than 1, make the value 1.
df['cholesterol']=[1 if x>1 else 0 for x in df['cholesterol']]
df['gluc']=[1 if x>1 else 0 for x in df['gluc']]
```

```python
# Draw Categorical Plot
def draw_cat_plot():
    # Create DataFrame for cat plot using `pd.melt` using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'.
    df_cat= df.melt(id_vars=['id','cardio'], value_vars=['cholesterol', 'gluc', 'smoke','alco','active','overweight'])
    fig=sns.catplot(x='variable', data=df_cat, kind='count', col='cardio', hue='value')
    fig.savefig('catplot.png')
    return fig
draw_cat_plot()
```

```python
#Clean the data. Filter out the patient segments that represent incorrect data
df_heat=df.loc[df['ap_lo'] <= df['ap_hi']]
df_heat=df.loc[df['height'] >= df['height'].quantile(0.025)]
df_heat=df.loc[df['height'] <= df['height'].quantile(0.975)]
df_heat=df.loc[df['weight'] >= df['weight'].quantile(0.025)]
df_heat=df.loc[df['weight'] <= df['weight'].quantile(0.975)]
df_heat=df_heat.drop(columns='BMI')
```

```python
# Draw Heat Map
def draw_heat_map():

    # Calculate the correlation matrix
    corr=df_heat.corr()

    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr))
    # Set up the matplotlib figure
    fig2, ax = plt.subplots(figsize=(7,7))
    # Draw the heatmap with 'sns.heatmap()'
    fig2 = sns.heatmap(corr.round(1), mask=mask, annot=True, linewidth=1,square=True)
    plt.savefig('heatmap.png')
    return fig2
draw_heat_map()
```
